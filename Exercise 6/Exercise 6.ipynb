## Determining the features that discriminate edibility the most

```python
# Imports
import pandas as pd
import numpy as np
import pandas.plotting

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Read dataset to dataFrame
dataFrame = pd.read_csv("agaricus-lepiota.data")
```

```python
# Display dataFrame
dataFrame.head()
```

```python
dataFrame.describe()
```

```python
# Createie dummie data and sperate out edibility column

x = pd.get_dummies(dataFrame.drop('edibility', 1))
y = pd.get_dummies(dataFrame.edibility)
```

```python
# Visualize the new dataFrames

x.head()
```

```python
y.head() # edibility
```

```python
skb = SelectKBest(chi2, k=10) # initialize selectKBest
skb.fit(x, y) # Pick k best from x based on y

x_transformed = skb.transform(x)
```

```python
mostDiscrimitiveFeatures = np.array(x.columns)[skb.get_support(indices = True)]
```

```python
for feature in mostDiscrimitiveFeatures:
    print(feature)
```

* Use principal components analysis to construct a reduced space. Which combination of features explain the most variance in the dataset?

## Which combination of features explain the most cariance in the dataset

```python
# Imports

from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
amount = 5
pca = PCA(n_components = amount)
pca.fit(x)

pca_x = pca.transform(x)
pca_x.shape

pcaFrame = pd.DataFrame(pca.components_, columns = x.columns)

for num in range(amount):
    print(++num)
    print(pcaFrame.T[num].nlargest(10))
    print() # Seperation
```
